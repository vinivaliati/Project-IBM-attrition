```python
df = df.drop(["EmployeeNumber"], axis = 1)
```

```python
colunas_categoricas_nao_ordenadas = [
    "BusinessTravel",
    "Department",
    "EducationField",
    "Gender",
    "JobRole",
    "MaritalStatus",
    "OverTime",
]

colunas_categoricas_ordenadas = [
    "Education",
    "EnvironmentSatisfaction",
    "JobSatisfaction",
    "JobInvolvement",
    "JobLevel",
    "PerformanceRating",
    "RelationshipSatisfaction",
    "StockOptionLevel",
    "WorkLifeBalance",
]

coluna_alvo = ["Attrition"]

colunas_numericas = [
    coluna for coluna in df.columns if coluna  not in (colunas_categoricas_nao_ordenadas + colunas_categoricas_ordenadas + coluna_alvo)
]
```

```python
colunas_numericas
```

```python
fig, axs = plt.subplots(2,7, figsize=(20,5))

for ax, coluna in zip(axs.flatten(), colunas_numericas):
    sns.histplot(data=df, x=coluna, ax=ax, kde=True)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2,7, figsize=(20,10))

for ax, coluna in zip(axs.flatten(), colunas_numericas):
    sns.boxplot(data=df, x=coluna, ax=ax, showmeans=True, hue="Attrition")

plt.tight_layout()
plt.show()
```

```python
corr = df.select_dtypes("number").corr()
matriz = np.triu(corr)

fig, ax = plt.subplots(figsize=(12,12))

sns.heatmap(
    corr,
    mask=matriz,
    annot=True,
    fmt=".2f",
    ax=ax,
    annot_kws={"fontsize":8}
)

plt.show()
```
